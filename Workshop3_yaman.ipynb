```python
total_fare = df["Fare"].sum()
print(total_fare)
```

```python
fare_p1 = df[df["Pclass"] == 1]["Fare"].sum()
fare_p2 = df[df["Pclass"] == 2]["Fare"].sum()
fare_p3 = df[df["Pclass"] == 3]["Fare"].sum()

print(fare_p1)
print(fare_p2)
print(fare_p3)
```

```python
def age_group(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

df["age_group"] = df["Age"].apply(age_group)
df["age_group"]
```

```python
total_passengers = len(df)
total_passengers
```

```python
group_counts = df["age_group"].value_counts()
group_counts
```

```python
group_percent = (group_counts / total_passengers) * 100
group_percent
```
