# **2. Composite Development Score:**
    • Create a new metric called Composite Score using the formula:
    Composite Score = 0.30 × Life Expectancy Index + 0.30 × GNI per Capita Index}
    Here: Life Expectancy Index → "life expectancy" and GNI per Capita Index → "gross inc percap"

    • Rank South Asian countries based on Composite Score.

    • Plot the top 5 countries in a horizontal bar chart.
    
    • Compare the ranking of countries by Composite Score with their HDI ranking and discuss any
    differences.

```python
south_asia_df['Composite Score'] = (
    0.30 * south_asia_df['life_expectancy'] +
    0.30 * south_asia_df['gross_inc_percap']
)
```

```python
# Select latest year data
latest_year = south_asia_df['year'].max()
south_asia_latest = south_asia_df[south_asia_df['year'] == latest_year]

# Rank by Composite Score
composite_ranking = south_asia_latest.sort_values(
    'Composite Score', ascending=False
)

composite_ranking[['country', 'Composite Score', 'hdi']]
```

```python
top5 = composite_ranking.head(5)

plt.figure(figsize=(7,5))
plt.barh(top5['country'], top5['Composite Score'])
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.title("Top 5 South Asian Countries by Composite Development Score")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Rank by HDI
hdi_ranking = south_asia_latest.sort_values('hdi', ascending=False)

# Compare rankings
comparison = composite_ranking[['country', 'Composite Score']].merge(
    hdi_ranking[['country', 'hdi']],
    on='country'
)

comparison
```

The Composite Development Score ranks South Asian countries based on life expectancy and income indicators.
Countries with higher income and better health outcomes rank higher on the Composite Score.
Differences between Composite Score ranking and HDI ranking highlight the importance of education and inequality components, which are included in HDI but excluded from the Composite Score.

# 7. Outlier Detection:
    • Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using the 1.5 × IQR rule.
    • Create scatter plots highlighting outliers in a different color.
    • Discuss the significance of these outliers.

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers
```

```python
sa_hdi_outliers = detect_outliers_iqr(south_asia_2020_2022, 'hdi')
sa_gni_outliers = detect_outliers_iqr(south_asia_2020_2022, 'gross_inc_percap')

sa_hdi_outliers[['country', 'year', 'hdi']]
```
